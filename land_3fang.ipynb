# 函数：get_page_list得到索引页

```python
def get_page_list(source):
    page_list = []  
    
    html = etree.HTML(source)
        
    a_href = html.xpath(".//div[@id='divAspNetPager']/a")[-2].xpath("./@href")[0]

    page_num = int(re.search("(.*_)(\d+).html", a_href).group(2))
    url_base1 = "https://land.3fang.com"
    url_base2 = re.search("(.*_)(\d+).html", a_href).group(1)
    url_base = url_base1 + url_base2

    page_list = []
    for i in range(1, page_num + 1):
        url = url_base + str(i) + ".html"
        page_list.append(url)
    return page_list
```
